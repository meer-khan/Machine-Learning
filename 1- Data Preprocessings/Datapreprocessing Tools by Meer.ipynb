# **Taking care of Missing data:**

```python
from sklearn.impute import SimpleImputer
imputer = SimpleImputer(missing_values=np.nan , strategy = "mean")
imputer.fit(X[:,1:3])
X[:,1:3] = imputer.transform(X[:,1:3])
print(X)
```

# **Encoding categorical data**

### OneHotEncoding

we have 3 categories (France, Spain, Germany) in our dataset, we cannot apply Simple Encoding by assigning the France 1, Spain 2, Germany 3. Because machine learning model will think that Germany is greater than the France beacuse Germany has a greater value then France. Since All these three categorical values doesnot have any corelation between them so we need to encode it in a way in which machine learning model will not give weight more to any of one catergorical value. 
So, we have one hot encoding technique in which encoder create seperate columns for all the unique values in categorical data colum (in our case it is 3). and assign each value 0 or 1 in the respective columns. like if we have the row of France, then the vector of that row would be [1 , 0 , 0] representing that, France is 1, Spain and Germany is 0. 


Params: remainder = "passthrough" will make sure that the remaining columns of the dataset will remain in the dataset during the fit_transform, otherwise, ColumnTransformer class will apply the transformation on the given column and remove all other columns. 

if we donot apply np.array() method in below code, then ColumnTranformer object will already give us the numpy array, so we donot need to apply np.array(). but in the course, instructor was apply so i am applying it too

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
ct = ColumnTransformer(transformers=[("encoder",OneHotEncoder(), [0])], remainder = 'passthrough')
print(ct)
```

```python
xok = ct.fit_transform(X)
print(xok)
print(type(xok))
```

```python
X = np.array(ct.fit_transform(X))
print(X)
print(type(X))
```

### Label Encoding on Dependent column "Purchased"
