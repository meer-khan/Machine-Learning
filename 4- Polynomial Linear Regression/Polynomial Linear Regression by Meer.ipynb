```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

#Reading Dataset

```python
dataset = pd.read_csv("Position_Salaries.csv")
X = dataset.iloc[:,1:-1].values
y = dataset.iloc[:,-1].values
```

iloc[:,1] vs iloc[:,1:-1] 
type of both is Data frame and sequence 
dataframe respectivey, sequence dataframe, returns 2D array in numpy when we apply .values on it 
and sequence returns 1D array in numpy

```python
print(type(X))
```

```python
y
print(type(y))
```

# Splitting data into trainning and test

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,random_state = 0)
```

# Applying Linear Regression

```python
from sklearn.linear_model import LinearRegression
lr = LinearRegression()
lr.fit(X,y)
```

# Applying polynomial linear regression 

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree = 2)
X_poly = poly_reg.fit_transform(X)
linear_reg = LinearRegression()
linear_reg.fit(X_poly,y)
```

```python
print(X_poly)
```

# Visualizing Results of Linear Regression 

```python
plt.scatter(X,y, color = "red")
plt.plot(X, lr.predict(X), color="blue")
plt.title("Truth and bluff")
plt.xlabel("Levels")
plt.ylabel("Salaries")
plt.show()
```

# Visualizing Results of Polynomial Linear Regression

```python
plt.scatter(X,y, color = "red")
plt.plot(X, linear_reg.predict(X_poly), color="blue", label="Degree " + str(2), linewidth=2)
plt.title("Truth and bluff")
plt.xlabel("Levels")
plt.ylabel("Salaries")
plt.show()
```

# Making more smoother curve(Tunning) 

By changing the degree of polynomialfeatures

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg2 = PolynomialFeatures(degree = 10)
X_poly2 = poly_reg2.fit_transform(X)
linear_reg2 = LinearRegression()
linear_reg2.fit(X_poly2,y)
```

# Visualizing the results again

```python
plt.scatter(X,y, color = "red")
plt.plot(X, linear_reg2.predict(X_poly2), color="blue")
plt.title("Truth and bluff")
plt.xlabel("Levels")
plt.ylabel("Salaries")
plt.show()
```

# Visualizing in more smoother curve

```python
X_grid = np.arange(min(X),max(X),0.1)
print(X_grid)
```

```python
X_grid = X_grid.reshape(len(X_grid), 1)
print(type(X_grid))
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg10 = PolynomialFeatures(degree = 10)
linear_reg3 = LinearRegression()
linear_reg3.fit(poly_reg10,y)

plt.scatter(X,y, color = "red")
plt.plot(X, linear_reg3.predict(poly_reg10.fit_transform(X_grid)), color="blue")
plt.title("Truth and bluff")
plt.xlabel("Levels")
plt.ylabel("Salaries")
plt.show()
```
